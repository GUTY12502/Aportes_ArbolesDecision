```python
#importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
#Se carga el dataset y se muestran las primeras cinco filas
data = pd.read_csv("winequality-red.csv")
data.head()
```

```python
#Verificamos que se carg√≥ completamente.
data.shape
```

```python
#Mostramos las medidas estadisticas
data.describe()
```

```python
#Vemos los valores nulos
data.isna().sum()
```

```python
#Realizamos un histograma donde se verifica los elementos por cada clase
plt.hist(data.quality)
```

```python
#Comparamos la densidad y el grado de alcohol con la calidad de cada uno a trav√©s de un grafico de dispersi√≥n
sns.scatterplot(x = "density", y = "alcohol", hue = "quality", data = data)
```

```python
#Dividimos el dataset, dejando las columnas predictoras y en el traget o objetivo, dejamos el dato a evaluar/comparar, en este caso, la calidad
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
predictors
```

```python
target
```

```python
#Hallamos la correlacion entre las variables predictoras
corr = predictors.corr()
corr
```

```python
#Los valores cercanos a uno, nos indica una alta correlacion directa, de lo contrario los cercanos a menos uno, indica una correlacion inversa.
```

```python
#Realizamos un mapa de calor con la misma informacion de correlacion
sns.heatmap(corr)
plt.show()
```

```python
#En este caso visulizamos exactamente lo mismo que en el anterior, solo que para este caso, a mayor correlacion directa, su color sera mas claro,
#caso contrario seria si su correlacion es inversa, se reflejaria con un color mas oscuro
```

```python
#Dividimos los datos en dos grupos para empezar con el arbol de decision
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.21, random_state=13)
```

```python
#Creamos el arbol de decision
tree = DecisionTreeClassifier()
```

```python
#Con los datos train o entrenamiento, empezamos a darle intruccion e informacion al arbol
arbol = tree.fit(X_train, y_train)
```

```python
#Visualizamos el arbol de decision con su grafica resultante
plot_tree(arbol)
```

```python
#Hacemos algunas predicciones de los datos
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
#Examinamos la matriz de confusi√≥n
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['actual'], colnames=['predicciones'])
```

```python
#Analizamos la metrica, comparando los valores reales con las predicciones realizadas
accuracy = accuracy_score(y_test,predicciones)
accuracy
```
